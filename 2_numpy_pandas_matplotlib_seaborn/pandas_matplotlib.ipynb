## Groupby

```python
# grouping the data by some values/parameters
# this sum will sum all the values of 0 deaths (ie., sums all sno with zero deaths)
df.groupby(['Deaths']).sum().sort_values("Deaths",ascending=False)
# groups the data by Deaths; ascending = False means in decending order
```

```python
# this gives mean of all the states and sorts by mean deaths in decending order
df.groupby(['State/UnionTerritory']).mean().sort_values('Deaths',ascending=False)
```

```python
# seperates all the State/UnionTerritory that contains "T"
df.loc[df['State/UnionTerritory'].str.contains("T")]
```

```python
# seperate all data with more that 0 death
df[df['Deaths']>0]
```

```python
# this gives the data with atleast one death and one cured on a same day
df[(df['Deaths']>0) & (df['Cured']>0)]
```

```python
# this gives the data type for each columns
df.dtypes
```

```python
# gets rid of all the empty/ null data, drops along axis=0 
# delete rows with atleast one empty/null data
df.dropna()
```

```python
# if there are 2 null values then drop that row
df.dropna(axis=0,thresh=2)
```

```python
# fills the empty placeholders with string "empty"
df.fillna(value='empty')
```

```python
# fills the empty/null cured column value with mean cured value
df['Cured'] = df['Cured'].fillna(value = df['Cured'].mean())
```

# Matplotlib

```python
x = np.arange(0,5,0.5)
y = x**3 - 3*x**2
z = x
```

```python
plt.figure(figsize=(10,6),dpi=300)      # this line is to make the figure larger

#  (x axis data; y axis data; color and format of line; etc )
plt.plot(x,-y,'r--',linewidth=2,markersize=12)   # this plots the data
```

```python
x = np.arange(0,5,0.5)
y = x**3 - 3*x**2
z = -4.4*x

plt.figure(figsize=(10,6),dpi=300)
# here we r ploting 2 things (line and a curve) in a single plot
plt.plot(x,-y,'r--',linewidth=2,markersize=4)
plt.plot(x,z,label='z-line')
plt.title('out first graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend() # if this line is not present it label on the top right wont show up
plt.xticks([1,2,3,4,5])  # this will set the values of x axis
plt.show()  # if this is not there it would also print the address the of the plot
```

```python
# bar plot
label = ['a','b','c']
value = [1,4,2.3]
plt.bar(label,value)
```

```python
label = ['a','b','c']
value = [1,4,2.3]
bars = plt.bar(label,value) 
pattern = ['/','o','*']
# looping through patterns and setting it in bar
for b in bars:
    b.set_hatch(pattern.pop()) # hatching the bar plot
plt.show()
```

## Histogram

a diagram consisting of rectangles whose area is proportional to the frequency of a variable and whose width is equal to the class interval.

It is similar to a Bar Chart, but a histogram groups numbers into ranges .
The height of each bar shows how many fall into each range.

```python
plt.figure(figsize=(16,12),dpi=600)
sns.distplot(df['SepalWidthCm'])
```

A box plot consist of 5 things.

-Minimum
-First Quartile or 25%
-Median (Second Quartile) or 50%
-Third Quartile or 75%
-Maximum

```python
sns.boxplot(x='SepalWidthCm', y='Id',hue='Species',data=df)
```

```python
sns.boxplot(x='PetalLengthCm', y='Id',hue='Species',data=df)
```
